```python
#Bai 15
from math import *
C = 4.186 
Gia_dien = 8.9 
Dvi = 2.777e-7  

M = float(input("Nhập khối lượng nước : "))
TT = float(input("Nhập sự thay đổi nhiệt độ : "))

# Năng lượng cần thiết 
Q = M * C * TT
# Đổi sang kWh
Q_kWh = Q * Dvi
# Tính chi phí
cost = Q_kWh * Gia_dien

print("Năng lượng cần thiết:",Q)
print("Đổi sang kWh: ",Q_kWh)
print("Chi phí năng lượng: ",cost)
```

```python
#Bai 17
from math import *
T = float(input("Nhập nhiệt độ không khí : "))
G = float(input("Nhập tốc độ gió : "))

Chi_so_gio = 13.12 + 0.6215 * T - 11.37 * (G ** 0.16) + 0.3965 * T * (G ** 0.16)
Lam_tron = round(Chi_so_gio)

print(f"Chỉ số gió lạnh là: ",Lam_tron)
```
